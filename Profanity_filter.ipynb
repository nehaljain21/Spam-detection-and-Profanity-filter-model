```python
import re
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'@\w+|#', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
df['clean_comment'] = df['comment_text'].apply(clean_text)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x = df['clean_comment']
y = df['toxic']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42,stratify=y)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
profanity_vectorizer = TfidfVectorizer(max_features=100000, ngram_range=(1,2))
x_train_vec = profanity_vectorizer.fit_transform(x_train)
x_test_vec = profanity_vectorizer.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
profanity_model = MultinomialNB()
profanity_model.fit(x_train_vec, y_train)
```

```python
y_pred = profanity_model.predict(x_test_vec)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
from joblib import dump, load
dump(profanity_model, "profanity_model.joblib")
profanity_model = load("profanity_model.joblib")
dump(profanity_vectorizer, "profanity_vectorizer.joblib")
profanity_vectorizer = load("profanity_vectorizer.joblib")
```
